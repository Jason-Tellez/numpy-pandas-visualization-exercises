```python
import numpy as np
import matplotlib.pyplot as plt
import math as m
```

#### Title your chart "Big O Notation"
#### Label your x axis "Elements"
#### Label your y axis "Operations"
#### Label your curves or make a legend for the curves
#### Use LaTex notation where possible

```python
plt.figure(figsize=(10,7))

# Range for graphs
x1 = np.arange(0.1, 15, .1)
# range for factorial
x = range(0, 15)

# equations
plt.plot(x1, [m.log(n) for n in x1], c = 'red', label = '$O(log n)$')
plt.plot(x1, [n for n in x1], c = 'green', label = '$O(n)$')

plt.plot(x, [0 * n + 1 for n in x], c = 'rosybrown', label = '$O(1)$')
plt.plot(x, [n * m.log(n + 1) for n in x], c = 'blue', label = '$O(n log n)$')
plt.plot(x, [n * n for n in x], c = 'orange' , label = '$O(n^2)$')
plt.plot(x, [2 ** n for n in x], c = 'cyan', label = '$O(2^n)$')
plt.plot(x, [m.factorial(n) for n in x], c = 'black', label = '$O(n!)$')
plt.plot(x, [n ** n for n in x], c = 'magenta', label = '$O(n^n)$')

#plot customization
plt.ylim(0, 13)
plt.xlim(0, 13)
plt.legend(loc= 'upper left')
plt.xlabel('Elements')
plt.ylabel('Operations')
plt.title('Big O Notation')
plt.show()
```
