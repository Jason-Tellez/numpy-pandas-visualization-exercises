### 1. Use matplotlib to plot the following equation:
   #####  y = x<sup>2</sup> âˆ’ x + 2
#### You'll need to write the code that generates the x and y points.
#### Add an anotation for the point 0, 0, the origin.

```python
x = range(0,26)
y = [n ** 2 - n + 2 for n in x]
plt.annotate('Origin', xy=(0, 0), xytext=(2.5, 100),
             arrowprops={'facecolor': 'blue'})
plt.plot(x, y)
plt.show()
```

### 2. Create and label 4 separate charts for the following equations (choose a range for x that makes sense):

```python
x = range(0,26)
y = [n ** .5 for n in x]
plt.title('Square root of x')
plt.plot(x, y)
plt.show()
```

### y = x<sup>3</sup>

```python
x = range(0,11)
y = [n ** 3 for n in x]
plt.title('Cube of x')
plt.plot(x, y)
plt.show()
```

### y = 2<sup>x</sup>

```python
x = range(0,11)
y = [2 ** n for n in x]
plt.title('2 to the power of x')
plt.plot(x, y)
plt.show()
```

### y = 1 / (x + 1)

```python
x = range(0,26)
y = [1 / (n + 1) for n in x]
plt.title('1 over x + 1')
plt.plot(x, y)
plt.show()
```

### 4. Combine the figures you created in the last step into one figure where each of the 4 equations has a different color for the points. Be sure to include a legend and an appropriate title for the figure.

```python
x = range(0,4)
a = [n ** .5 for n in x]
b = [1 / (n + 1) for n in x]
y = [n ** 3 for n in x]
z = [2 ** n for n in x]

plt.plot(x, a, c = 'r', label='$ \sqrt{x} $')

plt.plot(x, b, c = 'g', label='1/(1+x)')

plt.plot(x, y, c = 'b', label='x^3')

plt.plot(x, z, c = 'orange', label='2^x')

plt.title('Graphs of X')
plt.legend(loc='upper left')

plt.show()
```
